```python
import numpy as np
import matplotlib.pyplot as plt
import math
from typing import Tuple
```

```python
x = np.linspace(-10, 10.01, 1000)
```

```python
def drow_graph(x: np.ndarray, y: np.ndarray, d: np.ndarray, function_name: str, y_lim: Tuple, x_lim: Tuple=(-8, 8), minmax: bool=True) -> None:
    """
    x, y, d를 그래프로 출력해주는 함수
    
    input
    =====
    x: x
    y: f(x)
    d: df(x)/dx
    function_name: function f(x)'s name. for graph title
    y_lim: y_axis range
    x_lim: x_axis range. default is (-8, 8)
    minmax: show max and min values. defalut is True
    """
    fig, ax = plt.subplots(1, 2, figsize=(16, 4))

    ax[0].plot(x, y)
    ax[0].set_title(f"{function_name}")

    ax[1].plot(x, d)
    ax[1].set_title(f"{function_name} (Derivative)")
    if minmax:
        ax[0].axhline(max(y), color="red", linestyle="--", zorder=1)
        ax[0].axhline(min(y), color="blue", linestyle="--", zorder=1)
        ax[1].axhline(max(d), color="red", linestyle="--", zorder=1, xmax=0.5)

    for i in range(2):
        ax[i].axvline(0, color="gray", linestyle="-", zorder=0)
        ax[i].axhline(0, color="gray", linestyle="-", zorder=0)
        ax[i].set_xlim(x_lim)
        ax[i].set_ylim(y_lim)
        ax[i].spines['top'].set_visible(False)
        ax[i].spines['right'].set_visible(False)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
y = sigmoid(x)
d = y*(1-y)
```

```python
drow_graph(x, y, d, "Sigmoid Function", (-0.1, 1.1))
```

```python
def tanh(x):
    return 2*sigmoid(2*x)-1
```

```python
y = tanh(x)
d = 1-y**2
```

```python
drow_graph(x, y, d, "tanh Function", (-1.1, 1.1))
```

```python
def ReLU(x):
    return np.where(x<0, 0, x)
```

```python
y = ReLU(x)
d = np.where(y>0, 1, 0)
```

```python
drow_graph(x, y, d, "ReLu Function", (-0.1, 1.1), (-1, 1), False)
```

```python
def LeakyReLU(x: np.ndarray, alpha: float=0.01) -> np.ndarray:
    return np.where(x<0, alpha*x, x)
```

```python
alpha = 0.1
y = LeakyReLU(x, alpha)
d = np.where(y>0, 1, alpha)
```

```python
drow_graph(x, y, d, "LeakyReLU Function", (-0.1, 1.1), (-1, 1), False)
```

```python
def ELU(x: np.ndarray, alpha: float=0.01) -> np.ndarray:
    return np.where(x<=0, alpha*(np.exp(x)-1), x)
```

```python
alpha = 0.5
y = ELU(x, alpha)
d = np.where(y>0, 1, alpha*x*np.exp(x))
```

```python
drow_graph(x, y, d, "ELU Function", (-1, 4), (-3, 3), False)
```
